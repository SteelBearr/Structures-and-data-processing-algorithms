### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
class Stack:
    def __init__(self):
        self.items = []


    def check_empty(self):
        return len(self.items) == 0


    def push(self, item):
        self.items.append(item)


    def pop(self):
        if (not self.check_empty()):
            return self.items.pop()
    


class Deque:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def add_rear(self, item):
        self.items.append(item)

    def add_front(self, item):
        self.items.insert(0, item)

    def remove_front(self):
        if (not self.is_empty()):
            return self.items.pop()

    def remove_rear(self):
        if (not self.is_empty()):
            return self.items.pop(0)
        
    def size(self):
        return len(self.items)
    
    def peek(self):
        return self.items[-1]
    
f = open("Lab4_files/1_inp.txt")
books = f.read()
f.close()

deq1 = Deque()
deq2 = Deque()

books = books.split("\n")
print(books)
for item in books:
    deq1.add_rear(item)
    
while not deq1.is_empty():
    a = deq1.remove_rear()
    while not deq2.is_empty() and deq2.peek() > a:
        deq1.add_rear(deq2.remove_front()) 
    deq2.add_front(a)
fo = open("Lab4_files/1_out.txt",'w')
while not deq2.is_empty():
    t = deq2.remove_rear()
    fo.write(t + "\n")
    print(t)
fo.close()
```

### Задание №2
Дек содержит последовательность символов для шифровки сообщений. Дан текстовый файл, содержащий зашифрованное сообщение. Пользуясь деком, расшифровать текст. 
Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
f = "Lab4_files/2_inp.txt"
fo = "Lab4_files/2_out.txt"

def encryption(f,fo):
    f = open(f)
    st = f.read()
    print(st)
    f.close()
    d = Deque()
    fo = open(fo, 'w')
    for i in range (len(st)-2,len(st)+len(st)-2,1):
        if (i<len(st)):
            d.add_rear(st[i])
        else:
            d.add_rear(st[i-len(st)])
    fo.write(str(d.items))
    fo.close() 
    return d.items
    
print(encryption(f,fo))
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
f = "Lab4_files/3_inp.txt"
fo = "Lab4_files/3_out.txt"
a = Stack()
b = Stack()
c = Stack()
f = open(f)
n = int(f.read())
for i in range(0, n):
    a.push(n-i)
print("A: ", a.items)
print("B: ", b.items)
print("C: ", c.items)
print("----------------------")


def pyramids(a, b, c, n):
    if n == 1:
        c.push(a.pop())
    else:
        pyramids(a, b, c, n - 1)
        pyramids(a, c, b, 1)
        pyramids(b, c, a, n - 1)

pyramids(a, b, c , n)
b.items = []
a.push(c.pop())
fo = open(fo,'w')
a.items[0:], c.items[:] = c.items[0:], a.items[:]
fo.write(str(a.items)+'\n')
fo.write(str(b.items)+'\n')
fo.write(str(c.items)+'\n')
fo.write("----------------------")
fo.close()
print("A: ", a.items)
print("B: ", b.items)
print("C: ", c.items)
print("----------------------")
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
f = "Lab4_files/4_inp.txt"
fo = "Lab4_files/4_out.txt"
def round_brackets(f,fo):
    f = open(f)
    st = f.read()
    print(st)
    f.close()
    stck = Stack()
    fo = open(fo, 'w')
    for i in st:
        if i == "(":
            stck.push(i)
        elif i == ")":
            if stck.pop() == False:
                fo.write("False")
                fo.close()
                return False
    fo.write(str(stck.check_empty()))
    fo.close()
    return stck.check_empty()

print(round_brackets(f,fo))
```

### Задание №5
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс квадратных скобок в тексте, используя дек.

```python
f = "Lab4_files/5_inp.txt"
fo = "Lab4_files/5_out.txt"
def square_brackets(f,fo):
    f = open(f)
    st = f.read()
    print(st)
    f.close()
    deq = Deque()
    fo = open(fo, 'w')
    for i in st:
        if i == "]":
            deq.add_rear(i)
        elif i == "[":
            deq.add_front(i)
    l = deq.remove_rear()
    r = deq.remove_front()
    while not deq.is_empty():
        l = deq.remove_rear()
        r = deq.remove_front()
    fo.write(str((l == "[") and (r == "]")))
    fo.close()
    return (l == "[") and (r == "]")

print(square_brackets(f,fo))
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
f = "Lab4_files/6_inp.txt"
fo = "Lab4_files/6_out.txt"
def order(f,fo):
    f = open(f)
    str = f.read()
    print(str)
    f.close()
    stck = Stack()
    stck2 = Stack()
    ans = ""
    for i in str:
        if (ord(i) < 58) and (ord(i) > 47):
            ans += i
        elif (ord(i.lower()) < 123) and (ord(i.lower()) > 96):
            stck.push(i)
        else:
            stck2.push(i)
    acc = ""
    while len(stck.items) > 0:
        acc = stck.pop() + acc
    ans += acc
    acc = ""
    while len(stck2.items) > 0:
        acc = stck2.pop() + acc
    ans += acc
    fo = open(fo,'w')
    fo.write(ans)
    fo.close()
    return ans

print(order(f,fo))
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
f = "Lab4_files/7_inp.txt"
fo = "Lab4_files/7_out.txt"
def neg_pos(f,fo):
    f = open(f)
    st = f.read()
    f.close()
    print(st)
    deqe = Deque()
    arr = st.split(' ')
    right = ""
    left = ""
    for i in arr:
        if int(i) >= 0:
            deqe.add_front(int(i))
        else:
            deqe.add_rear(int(i))
    while not deqe.is_empty():
        i = deqe.remove_front()
        if i < 0:
            left = str(i) + " "+ left
        else:
            right += str(i) + " "
    fo = open(fo, 'w')
    fo.write(left + right)
    print(left + right)

neg_pos(f,fo)
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
f = "Lab4_files/8_inp.txt"
fo = "Lab4_files/8_out.txt"
def rev(fl, flo):
    fl = open(fl)
    stck = Stack()
    for line in fl:
        print(line, end="")
        stck.push(line.replace('\n',""))
    fl.close()
    print("\n----------------")
    flo = open(flo, 'w')
    while not stck.check_empty():
        i = stck.pop()
        print(i)
        flo.write(i+'\n')
    flo.close()
rev(f,fo)
```
