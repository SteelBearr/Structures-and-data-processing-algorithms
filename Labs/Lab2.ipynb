```python
#Бинарный поиск
import random
import timeit

arr = []
for i in range(40):
    arr.append(random.randrange(0, 99, 1))
arr.sort()
for item in arr:
    print(item, end=' ')
print()

def bin_search(arr:list,element:int):
    start = 0
    end = len(arr)

    while (start <= end):
        mid = (start + end) // 2

        if element == arr[mid]:
            return mid

        if element < arr[mid]:
            end = mid - 1
        else:
            start = mid + 1
    return -1

def bin_delete(arr:list,element:int):
    left = arr[:bin_search(arr,element)]
    right = arr[bin_search(arr,element)+1:]
    return left + right

def bin_add(arr:list,element:int):
    start = 0
    end = len(arr)

    while (start <= end):
        mid = (start + end) // 2

        if (element < arr[mid]) and (element > arr[mid-1]):
            a = arr[:mid]
            a.append(element)
            return a + arr[mid:]

        if element < arr[mid]:
            end = mid - 1
        else:
            start = mid + 1

start_time = timeit.default_timer()
print(bin_search(arr,7))
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
start_time = timeit.default_timer()
try:
    print(arr.index(7))
except ValueError:
    print(-1)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
```

```python
#Бинарное дерево
class Tree:

    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data


    def tree_add(self, data):

        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Tree(data)
                else:
                    self.left.tree_add(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Tree(data)
                else:
                    self.right.tree_add(data)
        else:
            self.data = data


    def tree_search(self, elem):
        if elem < self.data:
            if self.left is None:
                return False
            return self.left.tree_search(elem)
        elif elem > self.data:
            if self.right is None:
                return False
            return self.right.tree_search(elem)
        else:
            return True


    def print_tree(self):
        if self.left:
            self.left.print_tree()
        print(self.data),
        if self.right:
            self.right.print_tree()


tree = Tree(random.randrange(0, 99, 1))
arr=[]
for i in range(19):
    a = random.randrange(0, 99, 1)
    tree.tree_add(a)
    arr.append(a)
tree.print_tree()
print()
start_time = timeit.default_timer()
print(tree.tree_search(7))
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
start_time = timeit.default_timer()
try:
    print(arr.index(7))
except ValueError:
    print(-1)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
```

```python
#Метод Фибоначчи
arr = []
for i in range(40):
    arr.append(random.randrange(0, 99, 1))
arr.sort()
for item in arr:
    print(item, end=' ')
print()

def fibonacci_search(arr, element):
    fibM_minus_2 = 0
    fibM_minus_1 = 1
    fibM = fibM_minus_1 + fibM_minus_2
    while (fibM < len(arr)):
        fibM_minus_2 = fibM_minus_1
        fibM_minus_1 = fibM
        fibM = fibM_minus_1 + fibM_minus_2
    index = -1
    while (fibM > 1):
        i = min(index + fibM_minus_2, (len(arr) - 1))
        if (arr[i] < element):
            fibM = fibM_minus_1
            fibM_minus_1 = fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
            index = i
        elif (arr[i] > element):
            fibM = fibM_minus_2
            fibM_minus_1 = fibM_minus_1 - fibM_minus_2
            fibM_minus_2 = fibM - fibM_minus_1
        else:
            return i
    if(fibM_minus_1 and index < (len(arr) - 1) and arr[index + 1] == element):
        return index+1
    return -1

start_time = timeit.default_timer()
print(fibonacci_search(arr,7))
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
start_time = timeit.default_timer()
try:
    print(arr.index(7))
except ValueError:
    print(-1)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
```

```python
#Рехэширование с помощью псевдослучайных чисел
import random


class Hash_class:
    keys = []
    hashes = []

    def __init__(self, arr: list):
        self.add_array(arr)

    def add_key(self, item, length=None):
        if (length == None):
            length = len(self.keys)

        self.check_key_exists(item)
        temp = hash(item) % length

        if self.check_coll(temp):
            temp = self.rand_rehash(item, length)

        self.hashes.append(temp)
        self.keys.append(item)

    def print_hashes(self):
        for i in range(len(self.keys)):
            print(self.hashes[i], self.keys[i])
        print()

    def add_array(self, arr: list):
        for item in arr:
            self.add_key(item, len(arr))

    def rand_rehash(self, key, length: int):
        possible_values = [i for i in range(length * 2)]
        random.shuffle(possible_values)
        while True:
            i = possible_values.pop()
            temp = hash(key) % length + i
            if not self.check_coll(temp):
                return temp

    def check_coll(self, hash_value):
        return hash_value in self.hashes

    def check_key_exists(self, key):
        if key in self.keys:
            print(key + " exists")


values = ['toaster', 'bread', 'cookie', 'lattice', 'potatoes', 'carrot', 'new']
hashes = Hash_class(arr=values)

hashes.print_hashes()
hashes.add_key('newProduct')
hashes.add_key('newAnotherProduct')
hashes.print_hashes()
hashes.add_key('newProduct')
```

```python
#Метод цепочек
class Chains_class:
    hashes = {}
    values = {}
    pointer = 0
    N = 10

    def __init__(self, arr: dict = None):
        if arr != None:
            self.add_array(arr)

    def add(self, key, value):
        key_hash = hash(key) % self.N
        if self.check_coll(key_hash):
            prev_ind = self.hashes[key_hash]
            self.values[self.pointer] = [key, value, prev_ind]
        else:
            self.values[self.pointer] = [key, value]

        self.hashes[key_hash] = self.pointer
        self.pointer += 1

    def add_array(self, arr: dict):
        for key, value in arr.items():
            self.add(key, value)

    def find(self, key):
        key_hash = hash(key) % self.N
        if not key_hash in self.hashes:
            print('Item not found')
            return -1
        index = self.hashes[key_hash]
        return self.recursive_find(key, index)

    def recursive_find(self, key, index):
        if self.values[index][0] == key:
            return self.values[index][1]

        if len(self.values[index]) < 3:
            print('Item not found')
            return -1

        return self.recursive_find(key, self.values[index][2])

    def print_hashes(self):
        print("index => [key, value, pervious]:")
        for key, value in self.values.items():
            print("{} => {}".format(key, value))

        print("\nhash => index:")
        for key, value in self.hashes.items():
            print("{} => {}".format(key, value))

    def check_coll(self, key_hash):
        return key_hash in self.hashes



arr = {"key1": "value1", "key2": "value2", "key3": "value3"}
chains = Chains_class(arr)
chains.print_hashes()

print("\nappend 'newKey'")
chains.add("newKey", "newValue")
chains.print_hashes()

print("\nfind 'key2'")
print(chains.find("key2"))

print("\nfind 'key'")
print(chains.find("key"))
```
