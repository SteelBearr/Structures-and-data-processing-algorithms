### Выполнение:

```python
import cmath
import timeit
from cmath import pi
from tkinter import *


def canv_draw(l):
    global ox, oy, x, y, canvas
    oy = y
    ox = x
    canvas.create_line(ox, oy, x + (l * cmath.cos(ang)).real, y - (l * cmath.sin(ang)).real)
    x += (l * cmath.cos(ang)).real
    y -= (l * cmath.sin(ang)).real


def koch_curve(l, n):
    global ang, time_table
    time_table[n] = timeit.default_timer()
    if n == 0:
        canv_draw(l)
    else:
        koch_curve(l, n - 1)
        ang += pi/3
        koch_curve(l, n - 1)
        ang -= 2*pi/3
        koch_curve(l, n - 1)
        ang += pi/3
        koch_curve(l, n - 1)


def gilbert_curve_a(l, n):
    global ang, time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        ang -= pi / 2
        gilbert_curve_b(l, n - 1)
        canv_draw(l)
        ang += pi / 2
        gilbert_curve_a(l, n - 1)
        canv_draw(l)
        gilbert_curve_a(l, n - 1)
        ang += pi / 2
        canv_draw(l)
        gilbert_curve_b(l, n - 1)
        ang -= pi / 2


def gilbert_curve_b(l, n):
    global ang, time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        ang += pi / 2
        gilbert_curve_a(l, n - 1)
        canv_draw(l)
        ang -= pi / 2
        gilbert_curve_b(l, n - 1)
        canv_draw(l)
        gilbert_curve_b(l, n - 1)
        ang -= pi / 2
        canv_draw(l)
        gilbert_curve_a(l, n - 1)
        ang += pi / 2


def serp_curve_draw(l, n):
    global ang
    if n > 0:
        canv_draw(l)
        ang += pi / 2
        serp_curve(l, n)
        canv_draw(l)
        ang += pi / 2
        canv_draw(l)
        ang += pi / 2
        serp_curve(l, n)
        canv_draw(l)


def serp_curve(l, n):
    global ang, time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        serp_curve(l, n - 1)
        canv_draw(l)
        ang -= pi / 2
        canv_draw(l)
        ang += pi / 2
        canv_draw(l)
        ang -= pi / 2
        serp_curve(l, n - 1)
        canv_draw(l)
        ang += pi / 2
        canv_draw(l)
        ang += pi / 2
        serp_curve(l, n - 1)
        canv_draw(l)
        ang -= pi / 2
        canv_draw(l)
        ang += pi / 2
        canv_draw(l)
        ang -= pi / 2
        serp_curve(l, n - 1)


def serp_triang_draw(l, n):
    global ang
    if n > 0:
        serp_triang_a(l, n)
        serp_triang_b(l, n)
        serp_triang_a(l, n)
        ang-=2*pi/3
        serp_triang_a(l, n)
        serp_triang_a(l, n)
        ang -= 2 * pi / 3
        serp_triang_a(l, n)
        serp_triang_a(l, n)


def serp_triang_a(l, n):
    if n > 0:
        serp_triang_a(l, n-1)
        serp_triang_a(l, n-1)
    else:
        canv_draw(l)


def serp_triang_b(l, n):
    global ang, time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        ang -= 2 * pi / 3
        serp_triang_a(l, n-1)
        serp_triang_b(l, n-1)
        serp_triang_a(l, n-1)
        ang += 2*pi/3
        serp_triang_a(l, n-1)
        serp_triang_b(l, n-1)
        serp_triang_a(l, n-1)
        ang += 2 * pi / 3
        serp_triang_a(l, n-1)
        serp_triang_b(l, n-1)
        serp_triang_a(l, n-1)
        ang -= 2 * pi / 3


def serp_carp_draw(l, n):
    global x, y
    canvas.create_rectangle(x-l/2, y+l/2, x+l/2, y-l/2)
    serp_carp(x-l/2, y+l/2, x+l/2, y-l/2, n)


def serp_carp(a, b, c, d, n):
    global time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        a1 = 2 * a / 3 + c / 3
        c1 = a / 3 + 2 * c / 3
        b1 = 2 * b / 3 + d / 3
        d1 = b / 3 + 2 * d / 3
        canvas.create_rectangle(a1, b1, c1, d1)
        serp_carp(a, b, a1, b1, n - 1)
        serp_carp(a1, b, c1, b1, n - 1)
        serp_carp(c1, b, c, b1, n - 1)
        serp_carp(a, b1, a1, d1, n - 1)
        serp_carp(c1, b1, c, d1, n - 1)
        serp_carp(a, d1, a1, d, n - 1)
        serp_carp(a1, d1, c1, d, n - 1)
        serp_carp(c1, d1, c, d, n - 1)


def pifagor_tree_l(l, n):
    global ang, time_table
    if n > 0:
        time_table[n] = timeit.default_timer()
        ang += pi / 2
        canv_draw(l)
        ang -= pi / 4
        pifagor_tree_l(l / cmath.sqrt(2), n - 1)
        ang -= pi / 4
        canv_draw(l)
        pifagor_tree_r(l / cmath.sqrt(2), n - 1)
        ang -= pi / 2
        canv_draw(l)
        ang -= pi / 2
        canv_draw(l)
        ang += pi


def pifagor_tree_r(l, n):
    global ang
    if n > 0:
        ang += pi / 4
        canv_draw(l)
        ang -= pi / 2
        pifagor_tree_r(l / cmath.sqrt(2), n - 1)
        ang += pi
        canv_draw(l)
        ang -= 3*pi/4
        pifagor_tree_l(l / cmath.sqrt(2), n - 1)
        ang += 5*pi/4
        canv_draw(l)
        ang += pi / 2
        canv_draw(l)
        ang += pi / 4


window = Tk()
window.title('Кривая Коха')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 590
oy = 10
x = 10
y = 590
ang = 0
n = 7
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
koch_curve(2, n)
print("Кривая Коха")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()


window = Tk()
window.title('Кривая Гильберта')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 0
oy = 0
x = 50
y = 50
ang = 0
n = 7
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
gilbert_curve_a(4, n)
print("Кривая Гильберта")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
time_table.pop(len(time_table)-1)
time_table.insert(0,0)
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()


window = Tk()
window.title('Кривая Серпинского')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 300
oy = 590
x = 300
y = 590
ang = 0
n = 6
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
serp_curve_draw(2, n)
print("Кривая Серпинского")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
time_table.pop(len(time_table)-1)
time_table.insert(0,0)
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()


window = Tk()
window.title('Салфетка Серпинского')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 50
oy = 100
x = 50
y = 100
ang = 0
n = 7
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
serp_triang_draw(2, 7)
print("Салфетка Серпинского")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
time_table.pop(len(time_table)-1)
time_table.insert(0,0)
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()


window = Tk()
window.title('Ковёр Серпинского')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 300
oy = 10
x = 300
y = 300
ang = 0
n = 5
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
serp_carp_draw(598, 5)
print("Ковёр Серпинского")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
time_table.pop(len(time_table)-1)
time_table.insert(0,0)
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()


window = Tk()
window.title('Дерево Пифагора')
canvas = Canvas(window, width=600, height=600, bg="white", cursor="pencil")
ox = 250
oy = 590
x = 250
y = 590
ang = 0
n = 12
start_time = timeit.default_timer()
time_table = [start_time] * (n+1)
pifagor_tree_l(100, 12)
print("Дерево Пифагора")
for i in range(len(time_table)):
    if i == 0:
        print("n", end=" ")
    print(i, end=" ")
    time_table[i] -= start_time
    time_table[i] *= 1000
print()
time_table.reverse()
time_table.pop(len(time_table)-1)
time_table.insert(0,0)
print("t", end=" ")
for i in time_table:
    print(i, end=" ")
print()
canvas.pack()
window.mainloop()
```
