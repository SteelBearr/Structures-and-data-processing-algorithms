### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
    import random
    import timeit
```

```python
user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()
mat = [[0 for j in range(int(user_n))] for i in range(int(user_m))]
for i in range(int(user_m)):
    st=""
    for j in range(int(user_n)):
        r = random.randrange(int(user_min_limit),int(user_max_limit)+1,1)
        st = st + str(r) + " "
        mat[i][j] = r
    print(st)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
matrix = copy.copy(mat)
start_time = timeit.default_timer()
for i in range(len(matrix)-1):
    matrix[i].sort()
matrix.sort()
#TODO
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)):
    st=""
    for j in range(len(matrix[i])):
        st = st + str(matrix[i][j]) + " "
    print(st)
```

```python
# Сортировка вставкой.
matrix = copy.copy(mat)
def sort2(m):
    for i in range(1,len(m)):
        k = i
        while (k>0) and (m[k-1]>m[k]):
            t = m[k-1]
            m[k-1] = m[k]
            m[k] = t
            k-=1
start_time = timeit.default_timer()
#TODO
for i in range(len(matrix)):
    sort2(matrix[i])
sort2(matrix)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)-1):
    st=""
    for j in range(len(matrix[i])-1):
        st = st + str(matrix[i][j]) + " "
    print(st)
```

```python
# Сортировка обменом. || Сортировка пузырьком.
matrix = copy.copy(mat)
def sort3(m):
    for i in range (len(m)-1):
        for j in range (1,len(m)):
            if m[j-1] > m[j]:
                t = m[j-1]
                m[j-1] = m[j]
                m[j] = t
                
start_time = timeit.default_timer()
for i in range(len(matrix)):
    sort3(matrix[i])
sort3(matrix)
#TODO
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)):
    st=""
    for j in range(len(matrix[i])):
        st = st + str(mat[i][j]) + " "
    print(st)
```

```python
# Сортировка Шелла.
matrix = copy.copy(mat)
def sort4(m):
    step = len(m)//2
    while step > 0:
        for i in range(step, len(m)):
            j = i
            delta = j - step
            while (delta >= 0) and (m[delta] > m[j]):
                m[delta], m[j] = m[j], m[delta]
                j = delta
                delta = j - step
        step //= 2
start_time = timeit.default_timer()
#TODO
for i in range(len(matrix)):
    sort4(matrix[i])
sort4(matrix)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)):
    st=""
    for j in range(len(matrix[i])):
        st = st + str(matrix[i][j]) + " "
    print(st)
```

```python
# Быстрая сортировка.
matrix = copy.copy(mat)
def sort5(m):
    if len(m) <= 1:
        return m
    else:
        p = m[len(m)//2]
        l = []
        b = []
        e = []
        for i in m:
            if i < p:
                l.append(i)
            elif i > p:
                b.append(i)
            else:
                e.append(i)
        return sort5(l) + e + sort5(b)
start_time = timeit.default_timer()
#TODO
for i in range(len(matrix)):
    matrix[i] = sort5(matrix[i])
matrix = sort5(matrix)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)):
    st=""
    for j in range(len(matrix[i])):
        st = st + str(matrix[i][j]) + " "
    print(st)
    
```

```python
# Турнирная сортировка.
from heapq import *
matrix = copy.copy(mat)
def sort6(m):
    h = []
    for i in m:
        heappush(h, i)
    for i in range(len(m)):
        m[i] = heappop(h)

start_time = timeit.default_timer()
#TODO
for i in range(len(matrix)):
    sort6(matrix[i])
sort6(matrix)
print("--- {0} ms ---".format((timeit.default_timer() - start_time)*1000))
for i in range(len(matrix)):
    st=""
    for j in range(len(matrix[i])):
        st = st + str(matrix[i][j]) + " "
    print(st)
```
