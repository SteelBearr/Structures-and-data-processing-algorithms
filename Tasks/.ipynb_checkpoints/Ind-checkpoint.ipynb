```python
def swap(a, i, j):
    s = a[i]
    a[i] = a[j]
    a[j] = s


def next_set(a, n):
    j = n - 2
    while (j != -1) and (a[j] >= a[j + 1]):
        j -= 1
    if j == -1:
        return False
    k = n - 1
    while a[j] >= a[k]:
        k -= 1
    swap(a, j, k)
    l = j + 1
    r = n - 1
    while l<r:
        l += 1
        r -= 1
        swap(a, l, r)
    return True


a = input().split(" ")
a = [int(item) for item in a]
print(a)
while next_set(a, len(a)):
    print(a)
```
