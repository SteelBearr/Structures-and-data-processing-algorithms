```python
nums = [20, 39, 451, 1, 9, 8, 21]
a = nums
for j in range(0, len(a)-1):
    for i in range(0, len(a) - 1):
            if int(str(a[i+1])[0]) > int(str(a[i])[0]):
                t = a[i+1]
                a[i + 1] = a[i]
                a[i] = t
            for k in range(1, min(len(str(a[i + 1])), len(str(a[i])))):
                if (int(str(a[i+1])[0]) == int(str(a[i])[0])) & (int(str(a[i + 1])[k]) >= int(str(a[i])[k])):
                    t = a[i + 1]
                    a[i + 1] = a[i]
                    a[i] = t
ans = ""
for i in range(0, len(a)):
  ans += str(a[i])
print(ans)
```
